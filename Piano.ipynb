# Piano

Se implementa un martillo semicircular de diámetro 4mm (a lo más la superficie de contacto será de 4 mm)

```julia
function traslador(puntos, centro)
    nuevosx = puntos[1] + centro[1]
    nuevosy = puntos[2] + centro[2]
    nuevosx,nuevosy
end
```

```julia
function martillo(diametro,Δx)
    parte = Δx/(diametro)
    M=int(1/parte)
    y = [1:M+1]*0.0
    equis = [-diametro/2:Δx:diametro/2]

    for i in 1:M
        y[i] = sqrt((diametro/2)^2-(equis[i])^2)
    end
    p = equis, y
    nuevos = traslador(p,[0,-diametro/2])
    nuevos[1], nuevos[2]
end
```

```julia
function fuerzamartillo(Puntocuerda,indicepuntocuerda, puntomartillo,a,d ,Δx, constante)
    if a <= indicepuntocuerda <= d
        i = int(1+(indicepuntocuerda-a))
        z = (Puntocuerda-puntomartillo[i])
        if z >= 0.0 
            f=constante*z^3
        else 
            f = 0
        end
    else
        f = 0
    end
return f
end
```

```julia
function Piano(Condin,Posinicmartillo,β,L, T, μ, v0, K_fieltro, h_x, h_t, N) #T es la tensión de la cuerda,β la posición relativa del golpe, L la longitud de la cuerda, μ es la masa, h_ el tamaño de paso Δx o Δx, N el número de pasos de t
    Condin=hcat(Condin...)                                #v0, la velocidad inicial del martillo
    c=sqrt(T/μ) #velocidad de la cuerda
    r=c*h_t/h_x#constante
    M=length(Condin) #número de puntos y de la condición inicial
    L_h = β*L #La posición centrada del martillo respecto de uno de los cabos de la cuerda
    k_golpe = int(L_h/Δx) # Indice de la posción anterior
    radio = 2*Posinicmartillo[1][end] #radio del martillo
    pasos_radio = int(radio/Δx) #número de pasos en el intervalo con longitud radio
    masa = μ*Δx #La masa de cada uno de los elementos de la cuerda
    a = int(k_golpe-pasos_radio)
    d = int(k_golpe+pasos_radio)
    Y=zeros(N,M) 
    Y[1,:]=Condin
    Y[2,:]=Condin
    puntomartillo = Posinicmartillo
    velos = puntomartillo*0
        for l in 1:length(velos)
            velos[l]=v0
        end
      #v = Vector{Float64}[velos] #Arreglo para los vectores de velocidad
      #q = Vector{Float64}[puntomartillo] #Arreglo para los vectores de posición
      #t = Float64[0]
      #f = Vector{Float64}[]

    for n in 2:N-1 #El primer renglón que llenaré será el tercer renglón, aunque sea n=3, representará el primer paso de tiempo
        for i in 2:int(M-1)  #para no variar las condiciones de frontera Y[n,1]=Y[n,M]=0
            fuerz = fuerzamartillo(Y[n,i],i, puntomartillo,a,d,Δx, K_fieltro)
            Y[n+1,i]=2*(1-r^2)*Y[n,i]-Y[n-1,i] + (r^2)*(Y[n,i+1]+big(Y[n,i-1]))+(h_t^2/(μ*h_x))*fuerz#ojo, la numeración de renglones y columnas comienza en 1
             #EULER CROMER- SE IMPLEMENTA ASÍ PARA NO ESTAR LLAMANDOLO DESDE AFUERA UNA Y OTRA VEZ, TODA VEZ QUE LA FUNCION
            # VELOCIDAD ES -fuerz/m POR 3° LEY DE NEWTON, LO QUE SIENTA LA CUERDA DEBIDO AL MARTILLO SERÁ LO QUE SIENTE EL MARTILLO DEBIDO A LA COMPRESIÓN 
            # PROVOCADA POR LA DICHA COMPRESIÓN ¿Paré qué Euler-Cromer? Para calcular los nuevos puntos del martillo.
            
    #        push!(f,[fuerz])
            #push!(q, [puntomartillo])
            #push!(t,Δt*N)
            v_nueva = v0 - h_t*fuerz/masa
            q_nueva = puntomartillo + h_t*v_nueva
            puntomartillo = q_nueva #no interezan las velocidades del martillo, sino sus posiciones nuevas para hacer el cálculo.
            v0 = v_nueva
        end
    end
    Y #me devuelve la matriz de posiciones Y de la cuerda a todo tiempo
    #[t, hcat(q...)', hcat(v...)']
#f
#println(a)
#println(d)
end   
```

```julia
Δx=0.001; #Paso en x
L = 1     #Longitud de la cuerda
T = 100;  #Tensión
μ = 0.1;  #Densidad lineal
Δt=Δx/sqrt(T/μ); #Paso en t, considerando la velocidad de propagación en la cuerda
parte = Δx/L    #fracción del paso que representa Δx respecto de L, lo uso para un programa
N = int(0.05/Δt)        #Pasos en el tiempo para 1 segundo
β = 1/5         #Se le golpea a 1/5 de la longitud total
v0 = 5
CI=[0 for x in 0:Δx:L];
a=martillo(4,Δx)
K_fieltro = 100
```
